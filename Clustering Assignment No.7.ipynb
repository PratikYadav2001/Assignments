1. Airlines Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 300
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('EastWestAirline.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x = sc.fit_transform(df)
x
```

```python
# Hierarchical Clustering
```

```python
from scipy.cluster import hierarchy
```

```python
lk = hierarchy.linkage(x,method = 'complete')
dendrogram = hierarchy.dendrogram(lk)
```

```python
lk = hierarchy.linkage(x,method = 'average')
dendrogram = hierarchy.dendrogram(lk)
```

```python
# Model Building
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=5)
ypred = hc.fit_predict(x)
```

```python
df['cluster'] = ypred
```

```python
df
```

```python
df.groupby('cluster').agg('mean')
```

```python
df['cluster'].value_counts()
```

```python
sns.scatterplot(x = df['Balance'],y = df['Bonus_miles'],hue = df['cluster'],palette = 'rainbow')
```

```python
# Kmeans Clustering
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []


for i in range(1,19):
    km = KMeans(n_clusters = i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
    
```

```python
wcss
```

```python
plt.plot(range(1,19),wcss)
plt.axvline(x=7,linestyle='--',color='red')
plt.xlabel('K_values')
plt.ylabel('Wcss')
plt.title('Elbow Graph')
```

```python
km = KMeans(n_clusters=7)
ypred = km.fit_predict(x)
ypred
```

```python
df['cluster_km'] = ypred
df
```

```python
df.groupby('cluster_km').agg('mean')
```

```python
df.sort_values('cluster_km')
```

```python
df['cluster_km'].value_counts()
```

```python
# DBSCAN Clustering
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=3).fit(x)

d,index = neigh.kneighbors(x)
d
```

```python
distance = np.sort(d[:,1])
distance
```

```python
# K-distance Graph

plt.plot(distance)
plt.xlabel('Distance')
plt.ylabel('EPS')
plt.title(' K-distance Graph')
```

```python
db = DBSCAN(eps = 3,min_samples=3)
yp = db.fit_predict(x)
yp
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x,yp)
```

```python
df['cluster_db'] = yp
```

```python
df
```

```python
df.groupby('cluster_db').agg('mean')
```

```python
df[df['cluster_db'] == -1]
```

```python
df['cluster_db'].value_counts()
```
