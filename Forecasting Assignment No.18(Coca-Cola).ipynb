```python
# Sampling and Basic Transformation
```

```python
df = pd.read_excel('CocaCola_Sales_Rawdata.xlsx')
df
```

```python
df['Sales'].plot()
```

```python
# Splitting data
quarter = ['Q1','Q2','Q3','Q4']
```

```python
p = df['Quarter'][0]
p[0:2]
df['quarter'] = 0

for i in range(42):
    p = df['Quarter'][i]
    df['quarter'][i] = p[0:2]
    
    
df
```

```python
q_dummies = pd.DataFrame(pd.get_dummies(df['quarter']))
q_dummies
```

```python
df = pd.concat([df,q_dummies],axis=1)
df
```

```python
# Complete the dataset
df['t'] = np.arange(1,43)
df['t_square'] = np.square(df.t)
df['log_Sales'] = np.log(df.Sales)
df
```

```python
Train = df.head(35)
Test = df.tail(7)
```

```python
Train
```

```python
Test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#Linear Model
import statsmodels.formula.api as smf 

linear_model = smf.ols('Sales~t',data=Train).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(Test['t'])))
rmse_linear = np.sqrt(mean_squared_error(Test['Sales'], pred_linear))
rmse_linear
```

```python
#Exponential Model

Exp = smf.ols('log_Sales~t',data=Train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(Test['t'])))
rmse_Exp = np.sqrt(mean_squared_error(Test['Sales'], pred_Exp))
rmse_Exp
```

```python
#Quadratic Model

Quad = smf.ols('Sales~t+t_square',data=Train).fit()
pred_Quad = pd.Series(Quad.predict(Test[["t","t_square"]]))
rmse_Quad = np.sqrt(mean_squared_error(Test['Sales'],pred_Quad))
rmse_Quad
```

```python
#Additive seasonality Model
add_sea = smf.ols('Sales~Q1+Q2+Q3+Q4',data=Train).fit()
pred_add_sea = pd.Series(add_sea.predict(Test[['Q1','Q2','Q3','Q4']]))
rmse_add_sea = np.sqrt(mean_squared_error(Test['Sales'], pred_add_sea))
rmse_add_sea
```

```python
#Additive Seasonality Quadratic 

add_sea_Quad = smf.ols('Sales~t+t_square+Q1+Q2+Q3+Q4',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(Test[['Q1','Q2','Q3','Q4','t','t_square']]))
rmse_add_sea_quad = np.sqrt(mean_squared_error(Test['Sales'], pred_add_sea_quad))
rmse_add_sea_quad
```

```python
#Multiplicative Seasonality 

Mul_Add_sea = smf.ols('log_Sales~t+Q1+Q2+Q3+Q4',data = Train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(Test))
rmse_Mult_add_sea = np.sqrt(mean_squared_error(Test['Sales'], pred_Mult_add_sea))
rmse_Mult_add_sea 
```

```python
#Compare the results 

compare = {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea"]),"RMSE_Values":pd.Series([rmse_linear,rmse_Exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_add_sea])}
table_rmse=pd.DataFrame(compare)
table_rmse.sort_values(['RMSE_Values'])
```
